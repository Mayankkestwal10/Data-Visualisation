```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

%matplotlib inline
```

```python
tips = sns.load_dataset(name='tips')
```

```python
tips.head()
```

```python
sns.boxplot(x=tips['size']) 
```

```python
sns.boxplot(x='sex',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
#order
sns.boxplot(x='day',y='total_bill',data=tips,order=['Sat','Sun','Fri','Thur'])
```

```python
# orient = 'horizontal'
sns.boxplot(x='total_bill',y='day',data=tips,orient='horizontal',order=['Sat','Sun','Fri','Thur'])
```

```python
# boxplot + swarmplot
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='0.5')
```
