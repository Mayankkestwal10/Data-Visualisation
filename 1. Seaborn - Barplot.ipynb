```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#hue
sns.barplot(x='day',y='tip',hue='sex',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',hue='smoker',data=tips)
```

```python
# horizontal, palette
sns.barplot(x='total_bill',y='day',data=tips,palette='pastel')
```

```python
sns.barplot(x='day',y='tip',data=tips,palette='spring',order=['Sat','Sun','Thur','Fri'])
```

```python
# estimator

from statistics import median
sns.barplot(x='day',y='total_bill',data=tips,estimator=median)
```

<p>Error bars are graphical representations of data variability. They show the precision of a measurement. The bars usually represent standard deviation. It will the estimate standard deviation based on a sample.</p>

```python
# ci - confidence interval
sns.barplot(x='smoker',y='tip',data=tips,ci=99)
```

<p>A 95% ( default )confidence interval is a range of values that you can be 95% certain contains the true mean of the population. ... With large samples, you know that mean with much more precision than you do with a small sample, so the confidence interval is quite narrow when computed from a large sample.</p>

```python
sns.barplot(x='smoker',y='tip',data=tips,ci=50)
```

```python
# capsize
sns.barplot(x='day',y='tip',data=tips,palette='spring',capsize=0.3)
```

```python
sns.barplot(x='size',y='tip',data=tips,palette='autumn',capsize=0.15)
```
